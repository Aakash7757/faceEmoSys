```python
from keras.utils import to_categorical
from keras.preprocessing.image import load_img
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Input
import os
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
from sklearn.preprocessing import LabelEncoder
```

```python
trainDRI = 'images/train'
testDRI = 'images/test'
```

```python
def createdataframe(dir):
    imagePath = []
    labels = []
    for label in os.listdir(dir):
        for imagename in os.listdir(os.path.join(dir,label)):
            imagePath.append(os.path.join(dir,label,imagename))
            labels.append(label)
        print(label, "complited")
    return imagePath, labels
```

```python
train = pd.DataFrame()
train['image'], train['label'] = createdataframe(trainDRI)
print(train)
```

```python
test = pd.DataFrame()
test['image'], test['label'] = createdataframe(testDRI)
print(test)
```

```python
def EF(images):
    feature = []
    for image in tqdm(images):
        img = load_img(image, color_mode='grayscale')
        img = np.array(img)
        feature.append(img)
    feature = np.array(feature)
    feature = feature.reshape(len(feature),48,48,1)
    return feature
```

```python
trainF = EF(train['image'])
testF = EF(test['image'])
```

```python
le = LabelEncoder()
le.fit(train['label'])
```

```python
ytrain = le.transform(train['label'])
ytest = le.transform(test['label'])
```

```python
ytrain = to_categorical(ytrain, num_classes = 7)
test = to_categorical(ytest, num_classes = 7)
```

```python
xtrain = trainF/255.0
xtest = testF/255.0
```

```python
input_shape = (48, 48, 1)  # Example input shape for grayscale images of size 48x48
output_class = 7           # Example number of output classes for emotion recognition

# Initialize the model
model = Sequential()

# Add the Input layer
model.add(Input(shape=input_shape))

# Convolutional Layers
model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(256, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(512, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(512, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Flatten())

# Fully Connected Layers
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))

# Output Layer
model.add(Dense(7, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x = xtrain, y = ytrain, batch_size= 128, epochs = 100, validation_data = (xtest, ytest))
```
